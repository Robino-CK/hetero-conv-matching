```python
import pickle
with open(f'results/dblp_cca_new/0.7', 'rb') as fh:
    coarsener = pickle.load(fh) 
```

```python
sum_graph = coarsener.summarized_graph
sum_graph.nodes["author"].data.keys()
```

```python
X = sum_graph.nodes["author"].data["feat_pca"]
Y = sum_graph.nodes["author"].data["hauthortopaper"]
X_c, Y_c = coarsener.ccas["authortopaper"].transform(X,Y)

du = sum_graph.nodes["author"].data[f"deg_papertoauthor"]
print(du.unsqueeze(-1).shape)
R = X_c / du.unsqueeze(-1) + Y_c
print(R.shape)
```

```python
import torch
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
# Example data
data = R.cpu()
labels =sum_graph.nodes["author"].data["label"].cpu()

# Get the first two dimensions

data_np = data.numpy()

# Perform PCA
pca = PCA(n_components=2)
reduced = pca.fit_transform(data_np)  # Shape: (3652, 2)
sizes = [5, 5, 5, 5]
# Plot with color by label
plt.figure(figsize=(8, 6))
scatter = plt.scatter(reduced[:, 0], reduced[:, 1], c=labels.numpy(), cmap='tab10', alpha=0.7, sizes=sizes)
plt.xlabel("PCA Component 1")
plt.ylabel("PCA Component 2")
plt.title("PCA Projection (2D) Colored by Labels")
plt.grid(True)
plt.colorbar(scatter, label='Label')
plt.show()
```
